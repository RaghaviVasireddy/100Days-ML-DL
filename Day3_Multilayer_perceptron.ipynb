# Ovarian Cancer Detection using Multi Layer Perceptron (MLP)

## Project Overview
This project implements a **Machine Learning model using a Multi Layer Perceptron (MLP)** to classify ovarian cancer data. The dataset is preprocessed and normalized using **MinMaxScaler**, and then split into training and testing datasets.

The goal of this project is to demonstrate how neural networks can be applied for **medical data classification**.

---

## Dataset
Dataset used: `ovariantotal.csv`

The dataset contains multiple medical features and a target column:

- **TYPE** → Target variable representing cancer type/class.

```python
#importing the packages
import pandas as pd
import numpy as np
cancer=pd.read_csv('/content/drive/MyDrive/ovariantotal.csv')
cancer.head()
```

```python
cancer.tail()
```

```python
cancer.head(10)
```

```python
cancer.shape
```

```python
X=cancer.drop('TYPE',axis=1)
Y=cancer['TYPE']
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
X_scale=min_max_scaler.fit_transform(X)
print(X_scale)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scale,Y,test_size=0.20,random_state=0)
#random state split remain same
#test size 20% training size 80%
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(32,activation='relu',input_dim=49))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=32,epochs=15)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_pred=np.where(y_pred>0.5,1,0)
y_pred=y_pred.astype(int)
import numpy as np
np.column_stack((y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
