# Day 2 - Batch vs Online Learning

## Batch Learning
- Model is trained using the whole dataset at once.
- Training is done offline.
- If new data comes, retraining is needed.

## Online Learning
- Model learns step by step with new data.
- Training is continuous.
- Useful for streaming data.

## Main Difference
Batch: fit() on full data  
Online: partial_fit() on small data

```python
# Batch Learning Example

import numpy as np
from sklearn.linear_model import LinearRegression

# Full dataset
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X, y)   # batch training

print("Batch Prediction for 6:", model.predict([[6]]))
```

```python
# Online Learning Example

import numpy as np
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(max_iter=1, tol=None)

# Data comes one by one
X_stream = [[1],[2],[3],[4],[5]]
y_stream = [2,4,6,8,10]

for i in range(len(X_stream)):
    model.partial_fit([X_stream[i]], [y_stream[i]])
    print(f"Trained with sample {X_stream[i]}")

print("Online Prediction for 6:", model.predict([[6]]))
```
